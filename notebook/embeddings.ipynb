# Implementing the word embeddings with Tensorflow and Keras


```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

### Example sentences

```python

sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]

sent
```

## Define Vocabulary Size

```python
vocabulary_size = 10000
```

## One Hot representation of every words

```python
one_hot_representation =  [one_hot(input_text= words,
                                   n= vocabulary_size,
                                   lower = True,
                                   filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n')
                            for words in sent]

one_hot_representation
```

each word is assigned a specfic index in the vocabulary size of 10000, ex: the -> 8798

## Word Embedding representation

```python
from tensorflow.keras.layers import Embedding,Input
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
```

#### pad_sequence is used to set the maximum number of words in sentences as it is neccessary to train a model that each sentences has equal number of words

```python
sent_length = 8

embedded_docs = pad_sequences(sequences = one_hot_representation,
                              maxlen = sent_length,
                              padding ="pre")
embedded_docs
```

* before:

[8798, 7581, 9522, 2220],

[5799, 6775, 4816, 1279, 9480],


* after pre-padding:

[   0,    0,    0,    0, 8798, 7581, 9522, 2220],

[   0,    0,    0, 5799, 6775, 4816, 1279, 9480],


* after post-padding:

[8798, 7581, 9522, 2220,    0,    0,    0,    0],

[5799, 6775, 4816, 1279, 9480,    0,    0,    0],

## feature representation

```python
dim = 10
```

```python
model = Sequential()

model.add(Embedding(input_dim= vocabulary_size,
                    output_dim = dim,
                    input_length = sent_length))

model.compile(optimizer= "adam",
              loss = "mse")
```

```python
model.build(input_shape=(None, sent_length))
model.summary()
```
